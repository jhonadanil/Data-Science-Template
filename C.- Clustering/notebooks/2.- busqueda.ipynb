```python
import sys, os
sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from preprocessing.pipeline import codificar_genero
from src.config import SELECTED_FEATURES, BEST_PARAMS
```

### A.- Busqueda de hiperparametros

```python
df = pd.read_csv('../data/Mall_Customers.csv', index_col='CustomerID')

df, encoder = codificar_genero(df)
x = df[SELECTED_FEATURES]
```

```python
inertias = []
k_range = range(2, 15)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(x)
    inertias.append(kmeans.inertia_)
    labels = kmeans.labels_
    print(silhouette_score(x, labels))
```

```python
plt.plot(k_range, inertias, '-o')
plt.xlabel('Numero de clusters, k')
plt.ylabel('Inercia')
```

### B.- Entrenamiento

```python
kmeans = KMeans(**BEST_PARAMS)

kmeans.fit(x)
labels = kmeans.labels_
df['Cluster'] = labels
```

### C.- Evaluacion

```python
print(silhouette_score(df, labels))
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=labels, cmap='viridis')
plt.xlabel('Ingreso Anual (k$)')
plt.ylabel('Puntaje de Gasto (1-100)')
```
